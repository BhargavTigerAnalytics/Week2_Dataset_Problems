```python
sales = pd.read_excel('Week_2_Datasets/SalesData.xlsx')
```

## Q1. Find the least amount sale that was done for each item. 

```python
least_sales = sales.groupby('Item')['Sale_amt'].min().reset_index()
least_sales
```

## Q2. Finding total sales for each year and region across all items

```python
#converting the Order date column to date-time
sales['OrderDate'] = pd.to_datetime(sales['OrderDate'])

#making a new column for year
sales['year'] = sales['OrderDate'].dt.year
```

```python
#grouping the data by 'year' and 'region' columns
grouped_data = sales.groupby(['year', 'Region'])

#Sum the sales for each group
sales_sum = grouped_data['Sale_amt'].sum()
sales_sum = sales_sum.reset_index()

print(sales_sum)
```

## Q3. Create new column 'days_diff' with number of days difference between reference date passed and each order date

```python
import datetime
sales['OrderDate'] = pd.to_datetime(sales['OrderDate'])

#defining reference date
reference_date = datetime.datetime(2020, 1, 1)

#subtracting order date from ref date
difference = reference_date - sales['OrderDate']

#extracting no. of days
sales['days_diff'] = difference.dt.days

print(sales['days_diff'])
```

## Q4. Create a dataframe with two columns: 'manager', 'list_of_salesmen'. Column 'manager' will contain the unique managers present and column 'list_of_salesmen' will contain an array of all salesmen under each manager. 


```python
grouped = sales.groupby('Manager')

salesmen_lists = grouped['SalesMan'].unique()

salesmen_df = salesmen_lists.reset_index()

salesmen_df.columns = ['manager','list_of_salesmen']

salesmen_df
```

## Q5. For all regions find number of salesman and total sales. Return as a dataframe with three columns - Region, salesmen_count and total_sales

```python
# Grouping data by region
grouped_by_region = sales.groupby('Region')

#Count unique sales people per regiomn
salesmen_count_per_region = grouped_by_region['SalesMan'].unique().count()

#sales per region
sales_per_region = grouped_by_region['Sale_amt'].sum()

#combining both for new dataframe
region_stats = pd.DataFrame({
    'salesmen_count': salesmen_count_per_region,
    'total_sales': sales_per_region
})

#adding region as a column
region_stats = region_stats.reset_index()
region_stats
```

## Q6. Create a dataframe with total sales as percentage for each manager. Dataframe to contain manager and percent_sales 

```python
manager_grouped = sales.groupby('Manager')

manager_sales = manager_grouped['Sale_amt'].sum()
total_sales = manager_sales.sum()

sales_pct = (manager_sales / total_sales) * 100
sales_pct = sales_pct.reset_index()

sales_pct.columns = ['manager', 'percent_sales']


sales_pct

```

## Q14. Compute volume as (xyz) when depth is greater than 60. In case of depth less than 60 default volume to 8. 

```python
import numpy as np

diamonds['volume'] = np.where(
    diamonds['depth'] > 60, #condition
    diamonds['x'] * diamonds['y'] * diamonds['z'], #value if true
    8 #value if false
)

diamonds[['depth', 'x' , 'y' , 'z', 'volume']].head()
```

### Solving Q14 without using numpy

```python
volumes = []

for i in range(len(diamonds)):
    if diamonds.loc[i, 'depth'] > 60:
        v = diamonds.loc[i, 'x'] * diamonds.loc[i, 'y'] * diamonds.loc[i, 'z']
    else:
        v = 8
    volumes.append(v)

diamonds['volume'] = volumes

diamonds[['depth', 'x' , 'y' , 'z', 'volume']].head()
```

## Q15. Impute missing price values with mean. 

```python
mean_price = diamonds['price'].mean()

diamonds['price'] = diamonds['price'].fillna(mean_price)
diamonds['price'].isna().any()
```
